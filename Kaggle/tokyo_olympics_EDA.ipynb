```python
import pyspark

spark = pyspark.sql.SparkSession(pyspark.SparkContext())
```

```python
import pandas as pd

def excel_to_csv(name):
  pd.read_excel(name + '.xlsx').to_csv(name + '.csv',sep=',',index=False)

excel_to_csv('Athletes')
excel_to_csv('Coaches')
excel_to_csv('EntriesGender')
excel_to_csv('Medals')
excel_to_csv('Teams')
```

```python
athletes = pd.read_csv('Athletes.csv')
coaches = pd.read_csv('Coaches.csv')
entries_gender = pd.read_csv('EntriesGender.csv')
medals = pd.read_csv('Medals.csv')
teams = pd.read_csv('Teams.csv')
```

```python
medals.head()
```

```python
medals.sort_values(by='Rank by Total',inplace=True)
medals.head()
```

```python
import plotly.express as px

fig = px.bar(medals,x='Team/NOC',y=['Gold','Silver','Bronze'],title='Number of Medals per Country')
fig.show()
```

```python
coaches.head(20)
```

```python
spark.read.csv('Coaches.csv',header=True,inferSchema=True).createOrReplaceTempView('Coaches')
spark.sql("select * from Coaches limit 20").show()
```

```python
pd_coaches_query = spark.sql("select NOC,count(NOC) as Count from Coaches group by NOC order by Count desc").toPandas()
pd_coaches_query.head(7)
```

```python
fig = px.bar(pd_coaches_query,x='NOC',y='Count',color='Count',title='Number of Coaches per Country')
fig.show()
```

```python
athletes.head()
```

```python
spark.read.csv('Athletes.csv',header=True,inferSchema=True).createOrReplaceTempView('Athletes')
spark.sql("select * from Athletes order by NOC, Discipline").show(20)
```

```python
pd_athletes_query = spark.sql("select first(NOC) as NOC, first(Discipline) as Discipline, count(*) as Count from Athletes group by NOC, Discipline").toPandas()
```

```python
fig = px.bar(pd_athletes_query,x='NOC',y='Count', color='Discipline',title='Number of Athletes from each Country per Discipline')
fig.show()
```

```python
athletes['Discipline'].value_counts().head(10)
```

```python
pd_athletes_query = pd_athletes_query.loc[pd_athletes_query['Discipline'].isin(['Athletics','Swimming','Rowing'])]
pd_athletes_query.Discipline.value_counts()
```
