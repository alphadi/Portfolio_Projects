# Applying Classification Algorithm - Logistic Regression

```python
X = df.drop(columns='Class', axis=1)
Y = df['Class']
```

```python
print("Shape of predictor : ", X.shape)
print("\nShape of target : ", Y.shape)
```

```python
# Split train and test 70/30

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Fit the Model

model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
X_test = model.predict(X_train)
print("Accuracy score of Model on Training Data",accuracy_score(X_test, Y_train))
```
